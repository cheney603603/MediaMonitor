```python
import os
import re
```

```python
cur_path = os.getcwd()
paths = os.listdir(cur_path)
```

```python
paths.remove('data_pretreat.ipynb')
```

```python
res = {}
for p in paths:
    print('start treat: ',p)
    file_path = os.path.join(cur_path,p,'utf8')
    filenames = os.listdir(file_path)
    result = []
    for filename in filenames:
        with open(os.path.join(file_path,filename)) as file:
            lines = file.readlines()
            for line in lines:
                line.replace(' ','')
                line.replace('\t','')
                line = line=re.sub('[^(\u4e00-\u9fa5\u3002\uff1b\uff0c\uff1a\u201c\u201d\uff08\uff09\u3001\uff1f\u300a\u300b)]+','',line)
                if re.match('【.*】',line) or len(line)<5:
                    continue
                result.append(line)
    res[p.split('-')[1]]=result
```

```python
res
```

```python
os.mkdir('pretreat')
```

```python
for key in res.keys():
    with open(os.path.join(cur_path,'pretreat',key+'.txt'),'w') as file:
        for idx,line in enumerate(res[key]):
            if idx !=0:
                file.write('\n')
            file.write(line)
```
